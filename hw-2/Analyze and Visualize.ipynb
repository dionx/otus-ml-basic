### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
# Вообще, существует несколько формул вычисления среднего отклонения. В задаче явно не указана формула, поэтому
# использована простая формула вычисления среднего отклонения от среднего значения.

A = np.random.randint(1, 100, (7, 7))
print(A)

mean = A.mean()
abs_dev_mean = np.abs(A - mean)
mean_deviation = abs_dev_mean.mean()
print(f'Среднее отклонение: {round(mean_deviation, 2)}')

std = A.std()
print(f'Стандартное отклонение, вычисленное скучным образом: {round(std, 2)}')

D = ((A - mean) ** 2).mean()
std_2 = np.sqrt(D)
print(f'Стандартное отклонение, вычисленное вручную: {round(std_2, 2)}')

det = np.linalg.norm(A)
print(f'Определитель матрицы: {round(det, 2)}')
A_norm = A / det
print(f'Нормированная матрица:\n{A_norm}')

# Проверка что мы нашли именно нормализованную матрицу
assert(np.array_equal(A.T * A, A * A.T))
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
# Варинт 1. Не особо информативный
# sns.scatterplot(data=tips, x='total_bill', y='day', hue='day')

# Варинт 2. Громоздкий
# sns.displot(data=tips, x="total_bill", col="day", kde=True)

# Варинт 3. Сразу видно, что в пятницу ерунда какая-то
total_income = tips.groupby('day')['total_bill'].sum()
average_income = tips.groupby('day')['total_bill'].mean()

species = total_income.index.to_list()
x = np.arange(len(species))  # the label locations
width = 0.25 # the width of the bars

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

rect1 = ax1.bar(x, total_income, width, label="Total Income", color='coral')
rect2 = ax2.bar(x + width, average_income, width, label="Average Income", color='mediumblue')
# ax1.bar_label(rect1, padding=3)
# ax2.bar_label(rect2, padding=3)

ax1.set_ylabel('Total income ($)', color='coral')
ax2.set_ylabel('Average income ($)', color='mediumblue')
ax1.set_xlabel("DoW")
ax1.set_title('Income by DoW')
ax1.set_xticks(x + width / 2, species)

plt.show()

```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

### 1. О популярности заведения.

- Можно сделать вывод, что в рабочие дни в дневном меню есть в наличие бизнес-ланчи. В выходные дни кафе днем по всей видимости не работает вообще. 
Можно так же предположить, что приготовление еды для дневного меню происходит за счет остатков предыдущего дня, поэтому в пятницу используют днем только остатки еды с четверга, что отпугивает людей и они не хотят обедать по пятницам в этом заведении. Ну, или большинство трудяг работают по сокращенному рабочему дню и обедают дома.
Можно придумать еще разные гипотезы, почему в пятницу заведение не пользуется спросом.
- Заведение пользуется популярностью в вечернее время, особенно в weekend.

### 2. О счетах.
- Средний счет выше в выходные дни, что может быть связано с количеством посетителей, разнообразием меню и проведением досуга.
- Распределение суммы счета показывает, что большинство счетов находятся в диапазоне от 10 до 30 условных единиц.
- Мужчины весьма скупые и неохотно платят за женщин, что видно на сравнительно небольшом приросте суммы среднего чека в зависимости от пола.

### 3. О чаевых.
- Не совсем понятно в каком регионе была собрана статистика, скорее всего североамериканский континент, так как отсутствуют счета без чаевых, что немного искажает картину отношения посетителей к качеству заведения.
- Есть общий тренд, что вечером чаевых оставлют больше.
- Курящие более жадные на чаевые.
- Мужчины и женщины оставляют чаевые по-разному, мужчины немного более щедрые, чем женщины. Хотя это может быть следствием того, что у мужчин просто чуть большая сумма среднего чека
- Складывается ощущение, что если бы чаевые не являлись обязательными при оплате счета, то их бы почти никто и не платил.
