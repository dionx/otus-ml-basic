```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 204

A = np.random.rand(n, n)
# SVD 
U, s, VT = la.svd(A)
plt.plot(np.log(s))
```

$\tilde{A} = U[:,:r] S[:r, :r] V^{\top} [:r, :] $

```python
Sigma = np.diag(s)

Err = []
x = []
for r in range(2, n):
    x.append(r)
    A_approx = U[:, :r] @ Sigma[:r, :r] @ VT[:r, :]
    Err.append(la.norm(A - A_approx))  # Frobenius norm as default

plt.title("Распределение ошибки от ранга", y=1.05)
plt.plot(x, Err)
```
